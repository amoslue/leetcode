## How It Works (Step-by-Step)
Let’s sort [5, 3, 6, 2, 10]:

Find the smallest → 2, swap with first → [2, 3, 6, 5, 10]

Find smallest in rest → 3, already in place → [2, 3, 6, 5, 10]

Find smallest in rest → 5, swap with 6 → [2, 3, 5, 6, 10]

```python
def seletion_sort(nums):
    n = len(nums)
    for i in range(0, n):
        min_index = i
        for j in range(i+1, n):
            if nums[j]<nums[min_index]:
                min_index = j
        nums[i], nums[min_index] = nums[min_index], nums[i]
    return nums

nums = [5,2,6,4,8,56,1,45,6]
print(seletion_sort(nums))
```
