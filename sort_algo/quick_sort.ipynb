## How It Works (Step-by-Step)
Partition and return the index of pivot

recurse sorting until left>right
quick sort left part of pivot + quick sort right part of pivot



## Time Complexity
Best case: O(n log n)
Average case: O(n log n)
Worst case: O(nÂ²)

```python
def quick_sort(nums):
    def partition(low, high):
        pivot = nums[high]
        i = low - 1
        for j in range(low, high):
            if nums[j]<pivot:
                i+=1
                nums[i], nums[j] = nums[j], nums[i]
        
        nums[i+1], nums[high] = nums[high], nums[i+1]
        return i+1
    
    def sort_recur(low, high):
        if low<high:
            mid = partition(low, high)
            sort_recur(low, mid-1)
            sort_recur(mid+1, high)

    sort_recur(0, len(nums)-1)
    return nums

arr = [2,52,1,6,48,4,2,6,34,76]
print(quick_sort(arr))
```
