## Example
Sort [5, 3, 8, 4] using Bubble Sort:

First pass: [3, 5, 4, 8]

Second pass: [3, 4, 5, 8]

Third pass: [3, 4, 5, 8] (no swaps, done)

## Time Complexity
Best case (already sorted): O(n)

Average case: O(n²)

Worst case (reversed): O(n²)

Space complexity: O(1) (in-place)



```python
def bubble_sort(nums):
    n = len(nums)
    for i in range(n ,0, -1):
        for j in range(0, i-1):
            if nums[j+1]<nums[j]:
                swap = nums[j+1]
                nums[j+1] = nums[j]
                nums[j] = swap
    return nums

arr = [5, 1, 4, 2, 8, 12, 42, 35]
bubble_sort(arr)
print("Sorted array:", arr)

```
