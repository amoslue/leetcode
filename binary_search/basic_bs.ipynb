## lower_bound
1. right = n-1
2. while left<=right
3. if nums[mid] >= target
4. return left 
   
in this case, left equals to right+1 and it's the first index making nums[i] >= target

```python
def binary_search(nums, target):
    left = 0
    right = len(nums)-1
    while (left<=right):
        mid = (left+right) //2
        if nums[mid] >= target:
            right = mid-1
        else:
            left = mid+1
    return left

arr = [1,4,5,7,8,8,9,10]
target = 4
print(binary_search(arr, target))
```
